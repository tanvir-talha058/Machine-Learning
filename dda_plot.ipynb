```python
import matplotlib.pyplot as plt

# Starting and ending points
x1, y1 = 2, 3
x2, y2 = -2, -10

# Step 1: Calculate dx, dy
dx = x2 - x1
dy = y2 - y1

# Step 2: Determine steps
steps = max(abs(dx), abs(dy))

# Step 3: Calculate increments
x_inc = dx / steps
y_inc = dy / steps

# Step 4: Initialize starting point
x = x1
y = y1

# Step 5: Generate points using DDA
pixels = []
for i in range(int(steps) + 1):
    pixels.append((round(x), round(y)))
    x += x_inc
    y += y_inc

# Print pixel coordinates
print("Pixels on the line:")
for p in pixels:
    print(p)

# Step 6: Plot the points
x_coords = [p[0] for p in pixels]
y_coords = [p[1] for p in pixels]

plt.figure(figsize=(6, 6))
plt.scatter(x_coords, y_coords, color='red', s=80, label='Pixels')
plt.plot(x_coords, y_coords, linestyle='--', color='gray', label='DDA Path')
plt.title("DDA Line Drawing: P1(2,3) â†’ P2(-2,-10)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.gca().invert_yaxis()  # Optional: match computer graphics screen coordinates
plt.show()
```
