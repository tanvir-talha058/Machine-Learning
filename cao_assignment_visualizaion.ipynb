```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'Policy': ['FIFO', 'LRU', 'LFU'],
    'Hit Rate (%)': [65, 85, 77],
    'Execution Time (ms)': [170, 120, 140],
    'Energy (J)': [2.5, 1.8, 2.1],
    'IPC': [1.2, 1.6, 1.4]
}

# Create DataFrame
df = pd.DataFrame(data)
sns.set(style="whitegrid")

# Hit Rate
plt.figure(figsize=(6,4))
sns.barplot(x='Policy', y='Hit Rate (%)', data=df)
plt.title('Cache Hit Rate by Policy')
plt.savefig('hit_rate.png')

# Execution Time and Energy
plt.figure(figsize=(6,4))
df.plot(x='Policy', y=['Execution Time (ms)', 'Energy (J)'], kind='bar')
plt.title('Execution Time and Energy Comparison')
plt.savefig('execution_energy.png')

# IPC
plt.figure(figsize=(6,4))
sns.barplot(x='Policy', y='IPC', data=df)
plt.title('Instructions Per Cycle (IPC)')
plt.savefig('ipc_comparison.png')
```
